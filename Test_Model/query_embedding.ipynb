```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')

sentences = ["안녕하세요?", "한국어 문장 임베딩을 위한 버트 모델입니다."]
embeddings = model.encode(sentences)

print(embeddings)
```

```python
df['embedding'] = pd.Series([[]] * len(df)) # dummy
df['embedding'] = df['질문'].map(lambda x: list(model.encode(x)))
df.head()
```

```python
user_text = '내 개인 일정이 궁금해'
user_embedding = model.encode(user_text)
```

```python
df['distance'] = df['embedding'].map(lambda x: cosine_similarity([user_embedding], [x]).squeeze())
df.head()
```

```python
answer = df.loc[df['distance'].idxmax()]

print('의도 :', answer['의도'])
print('유사한 질문 :', answer['질문'])
print('챗봇 답변 :', answer['답변'])
print('유사도 :', answer['distance'])
```

```python
import re

def clean_text(texts):
    corpus = []
    for i in range(0, len(texts)):
        review = re.sub(r'[@%\\*=()/~#&\+á?\xc3\xa1\-\|\.\:\;\!\-\,\_\~\$\'\"]', '',str(texts[i])) #remove punctuation
        review = review.lower() #lower case
        review = re.sub(r'\s+', ' ', review) #remove extra space
        review = re.sub(r'<[^>]+>','',review) #remove Html tags
        review = re.sub(r'\s+', ' ', review) #remove spaces
        review = re.sub(r"^\s+", '', review) #remove space from start
        review = re.sub(r'\s+$', '', review) #remove space from the end
        corpus.append(review)
    return corpus
```

```python
import torch
import numpy as np
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('snunlp/KR-SBERT-V40K-klueNLI-augSTS')
```

```python
df['embedding_vector'] = df['질문'].map(lambda x : model.encode(x))
embedding_data = torch.tensor(df['embedding_vector'].tolist())
# torch.save(embedding_data, 'embedding_data.pt') 엔진 구현 사용용
```

```python
user_encode = model.encode(user_text)
user_tensor = torch.tensor(user_encode)

cos_sim = util.cos_sim(user_tensor, embedding_data)
```

```python
best_sim_idx = int(np.argmax(cos_sim))
selected_Q = df['질문'][best_sim_idx]
selected_Q
```

```python
slc_que_encode = model.encode(selected_Q)
score = np.dot(user_tensor, slc_que_encode) / (np.linalg.norm(user_tensor) * np.linalg.norm(slc_que_encode))
```

```python
print("질문 :",user_text)
print("선택된 질문 :",selected_Q)
print("가장 높은 코사인 유사도 idx :" ,best_sim_idx)
print("선택된 질문과의 유사도 :", score)
print("\n답변 :", df['답변'][best_sim_idx])
```
