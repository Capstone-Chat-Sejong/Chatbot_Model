```python
df['embedding'] = pd.Series([[]] * len(df)) # dummy
df['embedding'] = df['질문'].map(lambda x: list(model.encode(x)))
df.head()
```

```python
user_text = '내 개인 일정이 궁금해'
user_embedding = model.encode(user_text)
```

```python
df['distance'] = df['embedding'].map(lambda x: cosine_similarity([user_embedding], [x]).squeeze())
df.head()
```

```python
answer = df.loc[df['distance'].idxmax()]

print('의도 :', answer['의도'])
print('유사한 질문 :', answer['질문'])
print('챗봇 답변 :', answer['답변'])
print('유사도 :', answer['distance'])
```

```python
import re

def clean_text(texts):
    corpus = []
    for i in range(0, len(texts)):
        review = re.sub(r'[@%\\*=()/~#&\+á?\xc3\xa1\-\|\.\:\;\!\-\,\_\~\$\'\"]', '',str(texts[i])) #remove punctuation
        review = review.lower() #lower case
        review = re.sub(r'\s+', ' ', review) #remove extra space
        review = re.sub(r'<[^>]+>','',review) #remove Html tags
        review = re.sub(r'\s+', ' ', review) #remove spaces
        review = re.sub(r"^\s+", '', review) #remove space from start
        review = re.sub(r'\s+$', '', review) #remove space from the end
        corpus.append(review)
    return corpus
```

```python
%%writefile app.py

import streamlit as st
from streamlit_chat import message
import requests

st.set_page_config(
    page_title="Streamlit Chat - Demo",
    page_icon=":robot:"
)

#openai.api_key = st.secrets["api_secret"]
headers = {"Authorization": st.secrets['api_key']}

st.header("Sejong Smart Assistant")
st.markdown("[Github](https://github.com/Capstone-Chat-Sejong/Chatbot_Model)")

if 'generated' not in st.session_state:
    st.session_state['generated'] = []

if 'past' not in st.session_state:
    st.session_state['past'] = []

def generate_response(prompt):
    text = 'generated text'
    # cosine similarity // token
    message = text
    return message 

def get_text():
    input_text = st.text_input("You: ","Hello, how are you?", key="input")
    return input_text 


user_input = get_text()

if user_input:
    output = generate_response(user_input)
    # store the output 
    st.session_state.past.append(user_input)
    st.session_state.generated.append(output)

if st.session_state['generated']:
    for i in range(len(st.session_state['generated'])-1, -1, -1):
        message(st.session_state["generated"][i], key=str(i))
        message(st.session_state['past'][i], is_user=True, key=str(i) + '_user')


if __name__ == '__main__':
	main()
```
